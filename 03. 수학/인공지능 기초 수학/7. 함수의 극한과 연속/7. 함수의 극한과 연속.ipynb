```python
# 파이썬 sympy라이브러리에서 Limit, s, symbol 클래스를 호출
from sympy import Limit, S, Symbol
```

```python
# 변수 x를 생성하고 x가 한없이 0으로 가까워질 때의 1/x에 대한 극한값을 구한다

x=Symbol('x')
Limit(1/x, x , S.Infinity).doit()

# 첫 변수는 극한을 구하는 함수(1/x)
# 두 번째 변수는 X변수
# 세 번째 변수는 x변수가 다가가는 값

# 이 때 무한대를 정의하려고 s함수(S.Infinity)를 사용함
```

```python
# 우극한 값 구하기
Limit(1/x, x ,0).doit()
```

```python
# 좌극한 값 구하기
Limit(1/x, x ,0, dir='-').doit()

# 좌극한 값을 구할 때는 dir='-'를 사용함
```

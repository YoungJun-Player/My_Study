# 평균 변화율 구하기

```python
from sympy import symbols
```

```python
# 평균변화율을 구할 수 있는 함수를 정의

def average(a,b):
    m = max(a,b)    # a, b의 최댓값
    n = min(a,b)    # a, b의 최솟값
    x = symbols('x')    # 기호변수 x 선언

    fx = 2*x**2 + 4*x +7
    fb = fx.subs(x,m)   # 함수에 m대입
    fa = fx.subs(x,n)   # 함수에 n대입

    result = (fb-fa) / (m-n)
    return result

print(average(0,2))
```

# 미분계수(순간변화율) 구하기

```python
from sympy import Derivative, symbols # Derivative는 미분하는 함수
```

```python
# 평균변화율을 구할 수 있는 함수를 정의
x = symbols('x')
fx= 2*x**2 + 4*x + 7
```

```python
# fprime라는 Derivative 클래스의 객체를 생성한 후
# subs() 메서드를 사용하여 X = 3에서의 미분계수 f'(3)을 구한다

fprime = Derivative(fx, x).doit()   # x에 대한 미분
n = fprime.subs({x:3})

print(n)
```

# 미분하기

```python
import sympy as sym
from sympy import Derivative, symbols
```

```python
# x변수를 선언하고 diff() 함수를 사용하여 미분을 계산
x = symbols('x')
fx= sym.diff((2*x**3 + 3*x**2 +x+ 1),x)

print(fx)
```

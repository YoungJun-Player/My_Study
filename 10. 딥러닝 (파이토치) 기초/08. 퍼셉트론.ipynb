# 08. 퍼셉트론

출력이 0 또는 1인 작업을 의미하는 이진 분류 작업에 사용되는 간단한 모델

실제 뉴런이 신호를 전달하는 구조와 유사한 방식으로 구현

### 생물학적 신경망

가지돌기가 외부의 신경 자극을 받아 신경 세포체에서 가중 압력을 받아 신호 전달

전달되는 신호는 축삭을 통해 다른 신경 세포로 최종 신호를 전달

신경세포에서 다른 신경 세포로 전달할 때 시냅스라는 연결부위를 통해 신호를 전달

이때 전달되는 신호를 어느 정도의 세기로 전달할 지 결정하게 된다

### 퍼셉트론

생물학적 신경망과 유사하게 구성

가지돌기는 입력값(X)를 전달받는 역할

신경세포체는 입력값(x)를 토대로 특정 연산을 진행했을 때 임곗값보다 크면 전달하고 작으면 전달하지 않는다

계단함수를 적용해 결과를 반환한다

즉 입력값(X)와 노드의 가중치를 곱한 값을 모두 더했을 때 임곗값보다 크면 1을 출력하고 작으면 0을 출력한다

### 단층 퍼셉트론(SLP)


하나의 계층을 갖는 모델을 의미

입력을 통해 데이터가 전달되고 입력값(x)은 각각의 가중치와 함께 노드에 전달

전달된 입력값(x)과 가중치를 곱한 값이 활성화 함수에 전달

하나의 기울기로만 표현하기 어려운 경우에는 적용하기 어려움

### 다층 퍼셉트론(MLP)

단층 퍼셉트론을 여러 개 쌓아 은닉층을 생성
은닉층을 2개 이상 연결한다면 심층 신경망(DNN)이라고 부른다 

계층이 세 개 이상이라면 비교적 더 정확한 값을 찾을 수 있다 하지만 계층이 늘어날수록 갱신해야하는 가중치나 편향이 늘어난다 

최적의 가중치와 편향을 찾기 위해 많은 학습데이터와 연산량을 필요로 하게 된다


이진 분류 작업에서 여전히 사용되는 간단하고 효율적인 모델이지만 데이터의 복잡한 패턴을 학습할 수 없으며 선형으로 분리되지 않는 데이터를 분류할 수 없는 등 몇가지 제한 사항이 있다
